## Christmas Analysis

```python
## Importing Libraries

import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Loading data from MySQL Results

data = {
    'Product': ['Noise Canceling Buds', 'Bluetooth Speaker', 'Smart Coffee Mug', 
                'Silk Pajama Set', 'Leather Wallet', 'Electric Wine Opener', 'Gourmet Chocolate Box'],
    'Net Profit': [300.00, 225.00, 159.96, 135.00, 110.00, 85.00, 75.00],
    'Units Returned': [1, 0, 1, 1, 1, 1, 0]
}

df = pd.DataFrame(data)
```

```python
# Setup the festive theme
plt.figure(figsize=(10, 6))
colors = ['#2e7d32' if r == 0 else '#c62828' for r in df['Units Returned']] # Green for 0 returns, Red for returns

# Create the bar chart
bars = plt.bar(df['Product'], df['Net Profit'], color=colors)

# Customizing the looks
plt.title('Christmas Product Profitability: The Return Impact', fontsize=14, fontweight='bold', pad=20)
plt.ylabel('Net Profit ($)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add data labels on top of bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 5, f'${yval:.2f}', ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.show()
```
